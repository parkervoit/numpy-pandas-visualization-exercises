```python
# importing the libraries I need
import math
import matplotlib.pyplot as plt
```

```python
# this creates a list to operate on. .25 to get more resolution from the equations. 
x = [x * .25 for x in range(0 , 120)]

# factorials only work with whole numbers, so this list is for that
x1 = [x for x in range(0, 120)]

# These are the list comps to make lists to be plotted
o1 = [(0 * x) + 1 for x in x] # y = 0n+1

olog = [math.log(x) for x in x[1:]] # y = log(n)

n = [x for x in x] # y = n 

n2 = [x ** 2 for x in x] # y = n^2

nlog = [math.log(x) * x for x in x[1:]] # y = n * log(n)

o2n = [2 ** x for x in x] # 2^x

on1 = [math.factorial(x) for x in x1 ] # y = n! 

onn = [x ** x for x in x] # y = n^n

#create the figure
bon = plt.figure(figsize = (15,10))

# create x and y axis
plt.axhline(0,color='grey')
plt.axvline(0,color='black')

# name the figure
plt.title('Big O Notation', fontname = 'Chalkduster', fontsize = 40)

# label and assign tick marks for the axes, changed font to something bigger and also a more fun
plt.xlabel('Elements', fontname = 'Chalkduster', fontsize = 20)
plt.xticks([x * .5 for x in range(0,120)])
plt.ylabel('Operations', fontname = 'Chalkduster', fontsize = 20)
plt.yticks([y * .5 for y in range(-3,30)])

#plot functions in graph, with custom labels, markers, and linewidths
plt.plot(o1, label = '$O(1)$', marker = 'd', linewidth = 3 )
plt.plot(olog, label = '$Olog(n)$', marker = 's', linewidth = 3)
plt.plot(n, label = '$O(n)$', linewidth = 3)
plt.plot(nlog, label = '$O(n log n)$', ls = '--', linewidth = 3)
plt.plot(n2, label = '$O(n^2)$', ls =':', linewidth = 5)
plt.plot(o2n, label = '$O(2^b)$', ls = '-.', linewidth = 3)
plt.plot(on1, label = '$O(n!)$', color = 'yellow', linewidth = 3)
plt.plot(onn, label = '$O(n^n$)', ls = '--', color = 'red', linewidth = 3)

#place the legend in upper left, increasing size of text for readability
plt.legend(loc = 'upper left', fontsize = 15)

#limit length of axis to show negative y values, since log(x) = a negative number if ~3>x>0 
plt.ylim([-2,6])
plt.xlim([0,10])

#used a seaborn style to improve readability
plt.style.use('seaborn-deep')
```

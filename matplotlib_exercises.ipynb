```python
## imported matplotlib and math libraries to create graphs and do math
import matplotlib.pyplot as plt
import math
```

```python
#question 1 

# create x value set for computational purposes
data = list(range(30))

# list comprehension to create list of values to be plotted
y = [x ** 2 - x + 1 for x in data]

# xlim and ylim functions to make the lines start at origin 0,0
plt.text(0,0, '(0,0)', fontsize = 14, color = 'b')

# plot y values, change color to blue, and add a label to the line
plt.plot(y, c = 'orange', label = '$x^2 - x + 1$')
plt.show()
```

```python
#question 2

# use math function in the list comp to create list of values to be plotted
y0 = [math.sqrt(x) for x in data]

# plot y(0) to the graph
plt.plot(y0, label = '$sqrt$' )
plt.show()
```

```python
# x cubed list comp
y1 = [x ** 3 for x in data]

#limit and print
plt.plot(y1, label = '$x^2$')
plt.show()
```

```python
# 2 to the power of x list comp
y2 = [2 ** x for x in data]

# plot and show
plt.plot(y2, label = '$2^x$')
plt.show()
```

```python
# 1/(x+1) list comp to get working set
y3 = [1 / (x + 1) for x in data]

# plotted and print the graph
plt.plot(y3, label = '$1/(x+1)$')
plt.show()
```

```python
# Question 3
plt.figure(figsize = (10,6))

# grid is 1x4. 
# sub plot for sqrt
plt.subplot(221)

plt.plot(y0)
plt.title('$sqrt(x)$')


# subplot for cubed
plt.subplot(222)
plt.plot(y1)
plt.title('$x^3$')

# subplot for 2 to the power of x
plt.subplot(223)
plt.plot(y2)
plt.title('$2^x$')

#subplot for 1/(x+1)

plt.subplot(224)
plt.plot(y3)
plt.title('$1/(x+1)$')

# formatted the layout and gave the figure a name
plt.tight_layout()
plt.suptitle('All four functions')
plt.show()
```

```python
# plot the previous sets, adding figures, colors, and labels
plt.figure(figsize=(15,10))
plt.plot(y0, c = 'green', label = '$sqrt(x)$') #sqrt(x)
plt.plot(y1, c = 'red', label = '$x^3$')# x cubed
plt.plot(y2, c = 'blue', label = '$2^x$')# 2 to the power of x
plt.plot(y3, c = 'orange', label = '$ 1/(x+1)$') # 1/(x+1)
# add legend to upper right of screen 
plt.legend(loc = 'upper right')
# title the figure
plt.title("Question 3 equations")
# added limits to inprove readablility of the graph
plt.xlim([0, 20])
plt.ylim([0, 20])
plt.show()
```

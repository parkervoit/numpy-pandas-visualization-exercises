3. a) Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

3. b) Does whether or not a province is Catholic influence fertility? - we answered it a few ways.

- first with an lm plot, showing a trend line and a look at the clusters of data

```python
sns.lmplot(data = swiss_df, x = 'Catholic', y = 'Fertility',)
```

- from this we see that there is a weak correlation, and some very aggressive grouping between the catholics ( catholic >75) and the non catholics

```python
sns.boxplot(data = swiss_df, x = 'is_catholic', y = 'Fertility')
```

- the box plot shows that there isnt much overlap in the IQRs of catholics and non catholics when it comes to fertility, indicating that there is a correlation

```python
sns.heatmap(swiss_df.corr(), cmap = 'coolwarm_r', annot = True, vmax = 1, vmin = -1)
```

- this heatmap also shows that being catholic is positively correlated with fertility

I also decided to prove it statistically using scipy and an unpaired t test. I started by making a seperate dataframe of just catholics and another for just non catholics 

```python
catholics_df = swiss_df[swiss_df['is_catholic'] == True]
```

```python
noncat_df = swiss_df[swiss_df['is_catholic'] == False]
```

Then I used levenes test to determine if the variance of the population was different 

```python
sp.stats.levene(catholics_df['Fertility'], noncat_df['Fertility'])
```

The p value of .54 indicates that the variance is not statistically significant between the two groups. The next step is to run the t test

```python
sp.stats.ttest_ind(catholics_df['Fertility'], noncat_df['Fertility'] , equal_var=True)
```

The p value shows that the two groups are siginificantly different.

5. Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
sleep_df = data('sleepstudy')
```

```python
data('sleepstudy', show_doc= True)
```

```python
sleep_df
```

```python
# create figure to write to
plt.figure(figsize = (20,10))

# plot sleep_df data and also mean line using estimator. Make mean line transparent and thick to improve readability
sns.lineplot(data = sleep_df, x = 'Days', y = 'Reaction', 
            hue = 'Subject', legend = 'full',
            palette = 'pastel', linewidth = 3)
sns.lineplot(data = sleep_df, x = 'Days', y = 'Reaction', 
             color = 'blue', linewidth = 10, estimator = 'mean', alpha =.4)

# add a legend 
plt.legend(loc = 'upper left', fontsize = 'large')

# format title, ticks, and labels
plt.title('Reaction Times in a Sleep Deprivation Study', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.ylabel('Reaction', fontsize = 15)
plt.xlabel('Days', fontsize = 15)
```
